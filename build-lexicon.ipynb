```python
import pandas as pd
import json
```

```python
df_inset_positive = pd.read_csv("lexicon/InSet-positive.tsv", sep='\t')

df_inset_positive
```

```python
df_inset_negative = pd.read_csv("lexicon/InSet-negative.tsv", sep='\t')

df_inset_negative
```

```python
df_lexicon = pd.concat([df_inset_positive, df_inset_negative], ignore_index=True)
df_lexicon = df_lexicon.groupby(["word"]).sum().reset_index()

df_lexicon
```

```python
df_sentistrength = pd.read_csv("lexicon/sentiwords_id.txt", sep=":", names=["word", "weight"])

df_sentistrength
```

```python
df_sentistrength_emoticon = pd.read_csv("lexicon/sentistrength_id/emoticon_id.txt", sep="\s\|\s", names=["word", "weight"])

df_sentistrength_emoticon
```

```python
df_lexicon = pd.concat([df_lexicon, df_sentistrength, df_sentistrength_emoticon], ignore_index=True)

df_lexicon
```

```python
df_swear_words = pd.read_json("lexicon/swear-words.json").rename(columns={0:"word"})
df_swear_words["weight"] = -5

df_swear_words
```

```python
df_lexicon = pd.concat([df_lexicon, df_swear_words], ignore_index=True)

df_lexicon
```

```python
df_senti_tuned = pd.read_csv("lexicon/sentiwordnet_avg_dict_embedSG_oltw_tuned.txt", sep="\t", names=["word", "pos_score", "neg_score"])
df_senti_tuned["weight"] = (df_senti_tuned["pos_score"]-df_senti_tuned["neg_score"]) * 5.0


df_senti_tuned
```

```python
df_lexicon = pd.concat([df_lexicon, df_senti_tuned[["word", "weight"]]], ignore_index=True)

df_lexicon
```

```python
df_boosterwords_id = pd.read_csv("lexicon/sentistrength_id/boosterwords_id.txt", sep=":", names=["word", "weight"])

def map_value_booster(val):
    return "B_INCR" if val > 0 else "B_DECR"

df_boosterwords_id["booster"] = df_boosterwords_id["weight"].apply(map_value_booster)

df_boosterwords_id
```
